# 1. Read Dataset

```python
# Importing the pandas library to read the dataset
import pandas as pd

# read csv dataset with pd.read_csv() syntax
df = pd.read_csv("GRE-1300 Most commonly used words.csv")
```

# 2. Data Summary

```python
df.info()
```

# 3. Checking for Missing Data (Missing Value)

```python
# checking for missing value
df.isna().sum()
```

```python
# checking for statisticial summary
df.describe()
```

```python
# checking for statisticial summary
df['Frequency'].describe()
```

```python
# Resolve missing values
for column in df.columns:
    if df[column].dtype == 'object':
        # If the column is object type, fill it with mode
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # If the column is numeric type and the distribution is Skewness, fill it with the mode
        df[column].fillna(df[column].median(), inplace=True)
```

```python
# Recheck missing values
df.isna().sum()
```

```python
df.info()
```

# 4. Resolve Duplicate Data

```python
check_duplicate = df.duplicated().sum()

print(f"Number of duplicates data = {check_duplicate}")
```

```python
# Handling duplicates
df = df.drop_duplicates()
```

```python
# Check duplicates after handling
handle_duplicate = df.duplicated().sum()

print(f"Number of duplicates data = {handle_duplicate}")
```
